## File Input Setup & Inspection for SST
### This notebook is used to prep the two file inputs required for the runners/run_sst.py script
1. Watershed Boundary
2. Transposition Boundary

### Set the kernel of this notebook to stormhub-base after the devcontainer has been built

# SlamSim v2

```python
# Calculate areas in a projected CRS (EPSG:5070) for accurate measurements
transpo_projected = transpo_gdf.to_crs(epsg=5070)
valid_transpo_projected = valid_transpo_gdf.to_crs(epsg=5070)
huc4_projected = huc4_gdf.to_crs(epsg=5070)

# Calculate areas in square miles
transpo_area_sq_mi = transpo_projected.geometry.area.sum() / 1e6 * 0.386102
valid_transpo_area_sq_mi = valid_transpo_projected.geometry.area.sum() / 1e6 * 0.386102
huc4_area_sq_mi = huc4_projected.geometry.area.sum() / 1e6 * 0.386102

# Calculate ratios
original_ratio = transpo_area_sq_mi / huc4_area_sq_mi
valid_ratio = valid_transpo_area_sq_mi / huc4_area_sq_mi

print(f"HUC4 Watershed Area: {huc4_area_sq_mi:,.2f} sq mi")
print(f"Original Transposition Domain Area: {transpo_area_sq_mi:,.2f} sq mi")
print(f"Valid Transposition Domain Area: {valid_transpo_area_sq_mi:,.2f} sq mi")
print()
print(f"Original Transposition / HUC4 Ratio: {original_ratio:.2f}x")
print(f"Valid Transposition / HUC4 Ratio: {valid_ratio:.2f}x")
print(f"Valid / Original Transposition Ratio: {valid_transpo_area_sq_mi / transpo_area_sq_mi:.2%}")
```
